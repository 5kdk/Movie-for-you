```python
import pandas as pd
from konlpy.tag import Okt
import re
from tqdm import tqdm
```

```python
df = pd.read_csv('../crawling_data/reviews_2017.csv')
df.head()
```

```python
stopwords = pd.read_csv('../data/stopwords.csv', index_col=0)
stopwords
```

```python
movie_stopwords = ['영화', '배우', '감독']
stopwords_list = set(stopwords.stopword.to_list() + movie_stopwords)
```

```python
cleaned_sentences = []
okt = Okt()
for sentence in tqdm(df.reviews):
    sentence = re.sub('[^가-힣 ]', '', sentence)
    token = okt.pos(sentence, stem=True)
    words = [t[0] for t in token if t[1] in set(['Noun', 'Verb', 'Adjective']) and t[0] not in stopwords_list and len(t[0]) > 1]
    cleaned_sentence = " ".join(words)
    cleaned_sentences.append(cleaned_sentence)
df['cleaned_sentences'] = cleaned_sentences
df
```

```python
df = df[['titles', 'cleaned_sentences']]
df.to_csv('../data/cleaned_review_2017.csv')
df.info()
```
